```python
import io
from skimage import io
import  numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
image = io.imread("/home/pardus/İndirilenler/pandaa.jpg") 
def rgb_to_gray(image):
    R = (image[:, :, 0])
    G = (image[:, :, 1])
    B = (image[:, :, 2])
    
    r = np.floor((R *0.299)) 
    g = np.floor((G *0.587))
    b = np.floor((B *0.114))  
    print("r \n",r)
    print("g \n",g)
    print("b \n",b)
    gray_Image=(r+g+b)
    
    print("gray_Image \n",gray_Image)
    
    return gray_Image
gray_Image = rgb_to_gray(image) 

imagee = Image.fromarray(gray_Image.astype("uint8"),"L")
imagee.save("gray_Image_yether.png")
imagee.show()  #matplotlib görüntü işlemede kullanılmaz.
print("image size  \n",imagee.size)
```

```python
k=0
H=np.zeros((256))
#print(H)
while k<256: #while döngüsü ile k değerlerinmin 0'dan 256'ya kadar adım atması sağlanır. 
    H[k]=np.count_nonzero(gray_Image==k) #görüntüden gelen yoğunluk değerlerinin miktarının yazılması gerekir. 
    #Değerleri de H matrisine atarız.
    k=k+1
Intensity=np.arange(0,256,1)
#print(Intensity)
print(H)
plt.figure(figsize=(20,10))
plt.subplot(3,3,1)
plt.plot(Intensity,H,'k-')
plt.subplot(3,3,2)
plt.bar(Intensity,H,color='black',width=0.5) #çubuklardan oluşmasını istersek:
plt.xlabel("Intensity Value")
plt.ylabel("count")
plt.legend(['TOTAL'])
plt.show()
```

```python
#binary Image
thresholdValue = np.mean(gray_Image)
print(thresholdValue)
xDim, yDim = gray_Image.shape
 
print(xDim)
for i in range(xDim):
    for j in range(yDim):
        if (gray_Image[i][j]> thresholdValue):
            gray_Image[i][j] = 255
        else:
            gray_Image[i][j] = 0
#plt.imshow(gray_Image)
imag_e = Image.fromarray(gray_Image.astype("uint8"))
imag_e.save("imag_e.png")
imag_e.show()
```
